```python
import pandas as pd
```

```python
a = pd.read_csv('fatec_endereco_pessoa_fisica.csv')
a = a.dropna(subset=['cpf'])
a
```

```python
def converte(c):
    c = int(c)
    return c
```

```python
a['cpf'] = a['cpf'].apply(converte)
a
```

```python
def uf(h):
    h= str(h)

    if h[-3] == '0':
        return 'Rio Grande do Sul'
    if h[-3] == '1':
        return 'Distrito Federal–Goiás–Mato Grosso–Mato Grosso do Sul–Tocantins'
    if h[-3] == '2':
        return 'Pará–Amazonas–Acre–Amapá–Rondônia–Roraima'
    if h[-3] == '3':
        return 'Ceará–Maranhão–Piauí'
    if h[-3] == '4':
        return 'Pernambuco–Rio Grande do Norte–Paraíba–Alagoas'
    if h[-3] == '5':
        return 'Bahia–Sergipe'
    if h[-3] == '6':
        return 'Minas Gerais'
    if h[-3] == '7':
        return 'Rio de Janeiro–Espírito Santo'
    if h[-3] == '8':
        return 'São Paulo'
    if h[-3] == '9':
        return 'Paraná–Santa Catarina'
```

```python
#nulo = estado de origem do cpf
#des_estado = estado atual
#nom_cidade = cidade atual
#cpf, id_p_f, cidade atual, estado atual, estado de origem do cpf.
a['estado de origem do cpf'] = a['cpf'].apply(uf)
a
```

```python
q = a['estado atual'].value_counts()
q
```

```python
w = a['estado de origem do cpf'].value_counts()
w
```

```python
import matplotlib.pyplot as plt
```

```python
plt.barh(q.index, q.values)
plt.title('estado atual')
plt.xlabel('quantidade de CPFs')
plt.ylabel('Estados')
plt.savefig('estado atual.pdf')
plt.show()
```

```python
plt.barh(w.index, w.values)
plt.title('estado de origem do cpf')
plt.xlabel('quantidade de CPFs')
plt.ylabel('Estados')
plt.savefig('estado de origem do cpf.pdf')
plt.show()
```
